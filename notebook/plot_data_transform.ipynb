```python
import pandas as pd
```

```python
stock_file = "clean_stock_data.csv"
stonks = pd.read_csv(stock_file)
stonks.head()
```

```python
charger_file = "clean_charger_data.csv"
chrgrs = pd.read_csv(charger_file)
chrgrs.head()
```

```python
clean_data = chrgrs.loc[:, ["country","open_date"]]
clean_data
```

```python
clean_data=clean_data.rename(columns={"open_date":"date"})
```

```python
df_merge = pd.merge(clean_data, stonks, on='date')
```

```python
df_merge
```

```python
df_merge=df_merge.drop(['Unnamed: 0'],axis=1)
df_merge
```

```python
df_merge['index1'] = df_merge.index
df_merge
```

```python
df_merge['index1'] = df_merge.index1.values[::-1]
df_merge
```

```python
df_merge.to_csv('merged_data.csv')
```
