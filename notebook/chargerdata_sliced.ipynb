```python
# Import Dependencies
import pandas as pd
```

```python
# Read in csv data and assign it to a pandas dataframe
file = "Resources/Project2 data.csv"
df = pd.read_csv(file)
df.head()
```

```python
# Preview the datatypes of our columns
df.columns
```

```python
# Use the .loc() function to select only columns of interest
clean_data = df.loc[:, ["Country","City","GPS","Status","Open Date"]]
clean_data
```

```python
# Change column titles to more easily load data into SQL
clean_data=clean_data.rename(columns={"Country":"country","City":"city","GPS":"gps","Status":"status", "Open Date":"open_date"})
```

```python
# Filter so only the open Supercharger stations populate the dataframe
open_station= clean_data[clean_data['status']=='OPEN']
open_station
```

```python
# Convert the open date of the stations into appropriate format for use in SQL and JS/html.  
open_station['open_date']= pd.to_datetime(open_station['open_date'])
open_station
```

```python
# Split the 'GPS' columns into independent Latitude and Longitude columns 
open_station[['lat','long']] = open_station.gps.str.split(", ",expand=True)
open_station
```

```python
# Drop the now repetitive 'GPS' column
open_station=open_station.drop(['gps'],axis=1)
open_station
```

```python
# Upload our data into the pre-generated SQL database
from sqlalchemy import create_engine
connection = "postgresql://postgres:TRG19_sustain@localhost:6969/Project2_db"
engine = create_engine(connection)
```

```python
# Check connection to SQL database tables
engine.table_names()
```

```python
# Load data to SQL
open_station.to_sql(name='supercharged', con=engine, if_exists='append', index=False)
```

```python
# Save copy of data in HTML format for display on webpage  
save_html = open_station.to_html("clean_chargers.html", index=False)
```
